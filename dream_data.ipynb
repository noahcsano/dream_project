```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
career_dev = pd.read_csv(r"The Dream_ Measurement - Assessment & Evaluation - Career Development.csv")
comm_engage = pd.read_csv(r"The Dream_ Measurement - Assessment & Evaluation - Community Engagement.csv")
personal_dev = pd.read_csv(r"The Dream_ Measurement - Assessment & Evaluation - Personal Development, Misc..csv")
sa_perform = pd.read_csv(r"The Dream_ Measurement - Assessment & Evaluation - Student-Athlete Performance.csv")
```

```python
def combine_answer_choices(df):
    num_columns = len(df.columns)
    answer_choices = []
    
    for row in range(len(df)):
        for col in range(6, num_columns):
            item = df.iloc[row, col]
            if not pd.isnull(item):
                answer_choices.append(df.iloc[row, col])
        df["Anchors"][row] = answer_choices
        answer_choices = []
    df.drop(df.columns[7:], inplace = True, axis = 1)
    df.rename(columns={'Anchors': 'Answers'}, inplace=True)
```

```python
combine_answer_choices(career_dev)
career_dev["category"] = ["Career_Development" for i in range(len(career_dev))]
career_dev.set_index("category", drop=True)
career_dev.rename(columns={'What do we want to know?': 'Item'}, inplace=True)

combine_answer_choices(comm_engage)
comm_engage["category"] = ["Community_Engagement" for i in range(len(comm_engage))]
comm_engage.set_index("category", drop=True)

combine_answer_choices(personal_dev)
personal_dev["category"] = ["Personal_Development" for i in range(len(personal_dev))]
personal_dev.set_index("category", drop=True)

combine_answer_choices(sa_perform)
sa_perform["category"] = ["Performance" for i in range(len(sa_perform))]
sa_perform.set_index("category", drop=True)
True
```

```python
all_questions = pd.concat([career_dev, comm_engage, personal_dev, sa_perform]).set_index("category")
all_questions
```

```python
all_questions.to_csv('cleaned.csv')
```
